```python
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
iris['species']= label_encoder.fit_transform(iris['species'])
  
iris['species'].unique()
```

```python
# preparation of the data for traning
x=iris.iloc[:,1:5]
x.head()
```

```python
y=iris["species"]
y.head()
```

```python
x_train,x_test,y_tarin,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
#Training the model
Dtree=DecisionTreeClassifier(criterion="entropy",random_state=0)
model=Dtree.fit(x_train,y_train)
print("complete training")
```

```python
# prediction the model
y_predict=Dtree.predict(x_test)
print("The prediction value are :\n",y_predict)
print("\n the actual value are :\n",y_test.values.reshape(-1,45)[0])
```
